```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
student_df=pd.read_csv('Student_Performance.csv')
```

```python
student_df.head()
```

```python
student_df.rename(columns={'Hours Studied': 'HoursStudied'}, inplace=True)
student_df.rename(columns={'Previous Scores': 'PreviousScores'}, inplace=True)
student_df.rename(columns={'Extracurricular Activities': 'Activities'}, inplace=True)
student_df.rename(columns={'Sleep Hours': 'SleepHours'}, inplace=True)
student_df.rename(columns={'Sample Question Papers Practiced': 'Papers'}, inplace=True)
student_df.rename(columns={'Performance Index': 'Performance'}, inplace=True)
```

```python
student_df.head()
```

```python
student_df['Activities']=student_df['Activities'].map({'Yes':1,'No':0})
```

```python
student_df.head()
```

```python
student_df['Activities'].unique()
```

```python
inputs=student_df[['HoursStudied','PreviousScores','Activities','SleepHours','Papers']]
targets=student_df.Performance
```

```python
px.violin(student_df, x='HoursStudied', y='Performance')
```

```python
for col in inputs.columns:
    plt.scatter(student_df[col], targets)
    plt.xlabel(col)
    plt.ylabel("Performance")
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression().fit(inputs,targets)
```

```python
predictions=model.predict(inputs)
predictions
```

```python
from sklearn.metrics import root_mean_squared_error

loss=root_mean_squared_error(targets,predictions)
print(loss)
```
