```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
salaries_df=pd.read_csv('Salary_dataset.csv')
```

```python
inputs=salaries_df[['YearsExperience']]
targets=salaries_df.Salary
```

```python
inputs.head()
```

```python
inputs.shape
```

```python
targets.head()
```

```python
targets.shape
```

```python
plt.scatter(inputs,targets,marker='o',c='r')
plt.title("Salaries")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression().fit(inputs,targets)
```

```python
predictions=model.predict(inputs)
predictions
```

```python
targets.head(10)
```

```python
plt.plot(inputs,predictions,'r')
plt.scatter(inputs,targets)
plt.title("Salaries")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.legend(['Prediction','Actual'])
plt.show()
```

```python
from sklearn.metrics import root_mean_squared_error

loss=root_mean_squared_error(targets,predictions)
print(loss)
```

The parameters of the model are stored in the `coef_` and `intercept_` properties.
